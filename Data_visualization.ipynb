```python
# Step 1: Download the dataset
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows to understand the structure of the data
df.head()
```

```python
print(df.columns.tolist())   #to list the column names
```

```python
Create a bar chart summarizing the top 10 programming languages for the current year.
```

```python
# Split the semicolon-separated languages and count frequency
languages = df['LanguageHaveWorkedWith'].dropna().str.split(';')
language_counts = languages.explode().value_counts().head(10)

# Create bar chart
plt.figure()
language_counts.plot(kind='bar')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Respondents')
plt.title('Top 10 Programming Languages (Current Year)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
Create a bar or other relevant visualization summarizing the anticipated programming language trends
for the next year.
```

```python

# Split semicolon-separated languages and count them
future_langs = df['LanguageWantToWorkWith'].dropna().str.split(';')
future_counts = future_langs.explode().value_counts().head(10)  # top 10

# Plot
plt.figure()
future_counts.plot(kind='bar', color='teal')
plt.ylabel('Number of Respondents')
plt.xlabel('Programming Languages')
plt.title('Top 10 Anticipated Programming Languages for Next Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
Create a bar chart to summarize the top 10 databases currently in use.
```

```python
# Split semicolon-separated database names and count usage
databases = df['DatabaseHaveWorkedWith'].dropna().str.split(';')
db_counts = databases.explode().value_counts().head(10)

# Create bar chart
plt.figure()
db_counts.plot(kind='bar')
plt.xlabel('Databases')
plt.ylabel('Number of Respondents')
plt.title('Top 10 Databases Currently in Use')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Split semicolon-separated database names and count them
future_databases = df['DatabaseWantToWorkWith'].dropna().str.split(';')
db_future_counts = future_databases.explode().value_counts().head(10)

# Plot
plt.figure()
db_future_counts.plot(kind='bar', color='teal')
plt.xlabel('Databases')
plt.ylabel('Number of Respondents')
plt.title('Top 10 Databases anticipated for Future Use')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#First upload the csv file in the same path where this .ipynb file is there
!pip install openpyxl
import pandas as pd
import matplotlib.pyplot as plt
# Load the Excel file
file_path = "job-postings.xlsx"
df = pd.read_excel(file_path)

# Remove any repeated header rows if present
df = df[df["Technology"] != "Technology"]

# Convert JobPostings column to numeric
df["JobPostings"] = pd.to_numeric(df["JobPostings"], errors="coerce")

# Group by Technology and sort in descending order
job_counts = (
    df.groupby("Technology")["JobPostings"]
    .sum()
    .sort_values(ascending=False)
)

# Create bar chart
plt.figure()
plt.bar(job_counts.index, job_counts.values)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Technology")
plt.ylabel("Number of Job Postings")
plt.title("Job Postings by Technology (Descending Order)")
plt.tight_layout()
plt.show()
```

```python
#First upload the csv file in the same path where this .ipynb file is there
# Load the CSV file
file_path = "popular-languages.csv"
df = pd.read_csv(file_path)

# Display column names (optional check)
# print(df.columns)

# Convert Salary column to numeric (remove symbols if needed)
df["Average_Annual_Salary"] = (
    df["Average_Annual_Salary"]
    .astype(str)
    .str.replace(",", "")
    .str.replace("$", "")
    .astype(float)
)

# Sort data by Salary in descending order
df_sorted = df.sort_values(by="Average_Annual_Salary", ascending=False)

# Create bar chart
plt.figure()
plt.bar(df_sorted["Language"], df_sorted["Average_Annual_Salary"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Programming Language")
plt.ylabel("Average Salary")
plt.title("Average Salary by Programming Language (Descending Order)")
plt.tight_layout()
plt.show()
```
